# 입력 문자열에 따른 다음 문자열 예측 모델 만들기 
 - 예제) hello --> ello
  - h --> e
  - e --> l
  - l --> l
  - o --> ? 

## 0. install packages

```python
import torch
import torch.optim as optim
import pandas as pd
import numpy as np 
```

## 1. Data preperation

```python
sample = " if you want you"
```

```python
char_set = list(set(sample))
```

```python
char_set
```

```python
char_dic ={ char:idx  for idx, char in enumerate(char_set)}
```

```python
dic_size = len(char_dic)
hidden_size = len(char_dic)
learning_rate = 0.1
```

```python
sample_idx = [char_dic[c] for c in sample]
x_data = [sample_idx[:-1]]
x_one_hot = [np.eye(dic_size)[x] for x in x_data]
y_data = [sample_idx[1:]]
```

```python
x_one_hot
```

```python
x_data
```

```python
y_data
```

```python
X = torch.FloatTensor(x_one_hot)
```

```python
Y = torch.LongTensor(y_data)
```

## 3. Model implimentation

```python
torch.nn.RNN()
```

```python
input_size = dic_size
hidden_size = 10
rnn = torch.nn.RNN(input_size, hidden_size, batch_first=True)
#input = torch.randn(1, 5, input_size)
#output, hn = rnn(input)
```

```python
criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(rnn.parameters(), lr=0.001)
```

```python
output, _ = rnn(X)
print(X.size())
print(output.size())
```

```python
for i in range(50):
  rnn.train()
  optimizer.zero_grad()
  output, _ = rnn(X)
  #print(output, output.size())
  #print(Y, Y.size())
  loss = criterion(output.view(-1, 10), Y.view(-1))
  #print(output.view(-1, 10).size())
  #print(Y.view(-1).size())

  #print(output.data.numpy())
  #print(output.data.numpy().argmax(axis=2))
  result = output.data.numpy().argmax(axis=2)
  #print(result)
  result_str = ''.join([char_set[c] for c in np.squeeze(result)])
  print(i, "loss: ", loss.item(), "prediction: ", result, "true Y: ", y_data, "prediction str: ", result_str)
  loss.backward()
  optimizer.step()
```
