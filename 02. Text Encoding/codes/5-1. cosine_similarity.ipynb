# 1. 코사인 유사도(Cosine Similarity)

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np
```

```python
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
doc1 = np.array([0,1,1,1])
doc2 = np.array([1,0,1,1])
doc3 = np.array([2,0,2,2])
```

```python
print('Similarity between Documents 1 and 2 :',cos_sim(doc1, doc2))
print('Similarity between Documents 1 and 3 :',cos_sim(doc1, doc3))
print('Similarity between Documents 2 and 3 :',cos_sim(doc2, doc3))
```

# 2. Implementing a recommendation system using similarity

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Data Download: https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset
data = pd.read_csv('movies_metadata.csv', low_memory=False)
data.head(2)
```

```python
data = data.head(20000)
```

```python
print('Number of missing values in the overview column:',data['overview'].isnull().sum())
```

```python
data['overview'] = data['overview'].fillna('')
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data['overview'])
print('TF-IDF (shape) :',tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
print('Cosine similarity calculation results :',cosine_sim.shape)
```

```python
title_to_index = dict(zip(data['title'], data.index))
```

```python
idx = title_to_index['Father of the Bride Part II']
print(idx)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = title_to_index[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [idx[0] for idx in sim_scores]

    return data['title'].iloc[movie_indices]
```

```python
get_recommendations('The Dark Knight Rises')
```
