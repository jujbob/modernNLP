```python
import pandas as pd
import torch
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("AWE_1000_score.csv")
```

```python
df.head(2)
```

```python
vectorizer=TfidfVectorizer(
    min_df=2,
    max_features=10,
    sublinear_tf=True,
    lowercase=False,
    use_idf=True)
```

```python
vec_x_train = vectorizer.fit_transform(df["content"])
```

```python
vectorizer.vocabulary_
```

```python
vec_x_train[0].toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
vec_x_train.todense()[1].tolist()
```

```python
type(vec_x_train)
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model.fit(vec_x_train, df["score"])
predict = model.predict(vec_x_train)
```

```python
mean_squared_error(df["score"], predict)
```
