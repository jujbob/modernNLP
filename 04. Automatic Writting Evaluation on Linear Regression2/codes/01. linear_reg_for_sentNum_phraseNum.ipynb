### 왜 loss가 발산할까?
 - lr이 너무 높아서
 - sample이 너무 많아서 loss값이 애초에 크다
 

```python
class MyLinearModel(torch.nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = torch.nn.Linear(in_features=2, out_features= 1)

  def forward(self, x):
    return self.linear(x)
```

```python
myLin = MyLinearModel()
```
