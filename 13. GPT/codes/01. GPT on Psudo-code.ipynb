```python
import numpy as np
```

```python
#An example sentence :  Robot must obey the orders given it
vocab = ["Robot", "must", "obey", "the", "orders", "given", "it"]
ids = [0,1,2,3,4,5,6]
```

```python
def gpt(inputs):

  output = None
  #Trasformer Decoder blocks + Linear classifier
  if [0,1,2] == inputs:
    output = [0.0,   0.01,    0.02,   0.7,   0.0,  0.24, 0.03 ]
  elif [0,1,2,3] == inputs:
    output = [0.0,   0.01,    0.02,   0.1,   0.5,  0.24, 0.03 ]
  elif [0,1,2,3,4] == inputs:
    output = [0.0,   0.01,    0.02,   0.1,   0.0,  0.65, 0.03 ]
  elif [0,1,2,3,4,5] == inputs:
    output = [0.0,   0.01,    0.02,   0.1,   0.0,  0.05, 0.73 ]

  return output
```

```python
input_ids = [0,1,2]
```

```python
output = gpt(input_ids)
next_token_id = np.argmax(output)
next_token = vocab[next_token_id]
next_token
```

```python
def generate(inputs, n_tokens_to_generate):
    for _ in range(n_tokens_to_generate):
        out = gpt(inputs)
        next_id = np.argmax(out)
        inputs.append(int(next_id))
    return inputs[len(inputs) - n_tokens_to_generate :]
```

```python
input_ids = [0, 1, 2]
output_ids = generate(input_ids, 4)
output_tokens = [vocab[i] for i in output_ids]
```
